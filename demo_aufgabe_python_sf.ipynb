## Python entscheidet selbst durch *if*

Wir wollen nun lernen, wie man Wenn-Dann-Bedingungen in Python umsetzt. Dazu gibt es in unserem Kurs auch ein Video, das du dir [hier](https://playout.3qsdn.com/embed/e27aaa66-9b05-11ed-bf6a-3cecef385192) anschauen kannst.

Wir betrachten nun das folgende Szenario. Du bist in einem Team von Programmier:innen und ein Mobilfunkanbieter hat euch beauftragt, ein User-Interface für seine Kund:innen zu programmieren. Die erste Funktionalität, die ihr umsetzten sollt, ist die folgende: **Wenn** das Mobilfunkguthaben der Kund:innen weniger als 5 Euro beträgt, **dann** soll das Programm dies automatisch rückmelden und vorschlagen, dass das Guthaben aufgeladen werden soll. **Sonst** soll den Kund:innen gesagt werden, dass keine Aufladung nötig ist.

Hier muss also von Python eine **Bedingung** geprüft werden. Abhängig davon, ob diese erfüllt ist oder nicht, soll ein gewisser Code ausgeführt werden. Die gute Nachricht ist: Dies lässt sich in Python sehr leicht im sogenannten **if/else-Statements** umsetzen. 

Beginnen wir mit dem if-Teil. Wir definieren eine Variable ``mobile_balance`` und **wenn** diese einen Wert kleiner als 5 hat, **dann** lassen wir Python die folgende Ausgabe machen.

```
Your mobile balance is less than 5 EUR.
Maybe you should top up your balance.
```

Als Code lässt sich das wie folgt umsetzen. 


````python
mobile_balance = 4

if mobile_balance < 5:
    print('Your mobile balance is less than 5 EUR.')
    print('Maybe you should top up your balance.')
````

Schauen wir uns genau an, was da gerade passiert ist. Wir haben eine Variable `mobile_balance` mit dem Wert `4` erstellt. Dann kommt etwas Neues. Wir sehen, dass es zwei `print()`-Befehle gibt und beide ausgeführt wurden.

Ändere zunächst nur den Wert der Variable `mobile_balance`. Setze ihn zum Beispiel auf `11`. Was fällt dir auf? Genau, es wird nichts mehr gedruckt, das heißt keiner der beiden `print()`-Befehle wird ausgeführt.

Zwei Dinge sind für uns neu. Erstens taucht das Wort `if` und ein Doppelpunkt `:` auf. Zweitens sind die `print()`-Befehle nach rechts verschoben.

Wir müssen an dieser Stelle zunächst besprechen, was es mit der Verschiebung nach rechts auf sich hat. Wir sagen, dass der Code mit dem `print()`-Befehl *eingerückt* ist. Auch der englische Begriff ist wichtig - er lautet *indented*. Eine *Einrückung* ist eine *indentation*.

Diese Schreibweise könnte für dich ungewohnt sein. Wir verraten dir etwas: die Schreibweise ist in der Tat seltsam. Es gibt in fast allen Programmiersprachen das Wort `if`, aber die Einrückung ist eine Eigenart von Python. Leider haben wir eine schlechte Nachricht für dich: die Einrückung in Python muss zwingend genau richtig geschrieben werden und ist eine beliebte Fehlerquelle. Führe den folgenden Code aus:

````python
mobile_balance = 4

if mobile_balance < 5:
    print('Your mobile balance is less than 5 EUR.')
   print('Maybe you should top up your balance.')
````

Wir erhalten einen `IndentationError`. Es wird wahrscheinlich nicht das letzte Mal sein, dass du diesen Fehler siehst. Eine korrekte Einrückung besteht aus *vier Leerzeichen*.  Wirklich, es müssen *vier Leerzeichen* sein. Am besten du korrigierst den obenstehenden Code.

Nun haben wir eine gute Nachricht für dich: unsere Lernumgebung hilft dir mit der korrekten Einrückung. Wenn du einen Doppelpunkt schreibst und dann Enter drückst, wirst du sofort an die richtige Stelle bewegt. Dies solltest du nun üben.

## Python entscheidet zwischen zwei Codeblöcken mit *if*-*else*

Wenn man lediglich mit `if` Codeausführung kontrolliert, kann man nur angeben, dass Code ausgeführt wird, wenn eine Bedingung `True` ist. Bisher geschah bei `False` einfach gar nichts. Damit auch bei `False` etwas geschieht, erweitern wir das `if`-*statement* mit einem `else`-*statement*.

Beispielsweise könnte das Computerprogramm des oben genannten Handykunden bei genügend Guthaben den Text

```python
'No top-up necessary.'
```

ausgeben. Wie würde man diese Situation in Python darstellen?

1_01_08_pic2_if_else.png

````python
# customer data
mobile_balance = 4
bank_balance = 150

# automatic transfer algorithm
if mobile_balance < 5:
    mobile_balance = mobile_balance + 5
    bank_balance = bank_balance - 5
else:
    feedback_text = 'No top-up necessary.'
    print(feedback_text)

print(mobile_balance)
print(bank_balance)
````

Weil die Bedingung des `if`-*statements* `False` ist, wird der eingerückte Code von `if` übersprungen. Stattdessen wird der Code unter `else` ausgeführt. Setze den Wert von `mobile_balance` nun auf 11 EUR um zu sehen, dass die Ausgabe `No top-up necessary` bekommst.
